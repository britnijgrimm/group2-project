```python
# Dependencies and Setup
import pandas as pd
```

```python
# Files to load for appointment and city data
app_data = "Resources/processeddata.csv"
city_data = "Resources/brazil_cities_clean.csv"

# Read the appointment and city data and store into Pandas DataFrames
app_data_load_df = pd.read_csv(app_data)
city_data_load_df = pd.read_csv(city_data)

#Preview appointment data
app_data_load_df
```

```python
#Preview city data
city_data_load_df.head()
```

```python
#Select only city and lat/long data for city data
city_data_df = city_data_load_df[['CITY', 'STATE', 'LAT', 'LONG','ESTIMATED_POP']]

city_data_df.head()
```

```python
#Clean up column names for city data
city_data_df.rename(columns={'CITY': 'City', 'STATE': 'State', 
                             'LAT': 'Latitude', 'LONG': 'Longitude', 
                             'ESTIMATED_POP': 'Esitmated Population (2018)'}, inplace=True)

city_data_df.head()
```

```python
#Correct case on neighborhood app data
app_data_df = app_data_load_df

app_data_df['Neighbourhood'] = app_data_df['Neighbourhood'].str.title()
```
